```
vehicles = spark.sql("SELECT * FROM vehicle_process_csv")
vehicles.createOrReplaceTempView("vehicle_process_csv")
display(vehicles.select("*"))
```

```
#input a copy of vehicle table, in order to demostrate the join function
```

```
vehicles_copy = spark.sql("SELECT * FROM copy")
vehicles_copy.createOrReplaceTempView("copy")
display(vehicles_copy.select("*"))
```

```
vehicles.dtypes
```

```
vehicles_copy.dtypes
```

```
vehicles.describe().show()
```

```
display(vehicles.select("Vehicle_Type"))
```

```
display(vehicles.select("Left"))
```

```
#join two tables 
```

```
vehicles.join(vehicles_copy, vehicles.Photo_Name == vehicles_copy.name, 'outer').select(vehicles.Photo_Name, vehicles_copy.type).collect()
```

```
#sort Left column based on descending 
```

```
display(vehicles.sort(vehicles.Left.desc()).collect())
```

```
#aggregate fuction: average: show four numeric columns average
```

```
display(vehicles.groupBy().avg().collect())
```

```
#aggregate fuction: average: show Left column average
```

```
display(vehicles.groupBy().avg("Left").collect())
```

```
#get Left count based on count ascending 
```

```
display(vehicles.groupBy(vehicles.Left).count().collect())
```

```
# get count based on Left value ascending 
```

```
display(sorted(vehicles.groupBy(vehicles.Left).count().collect()))
```

```
# get max Left value
```

```
vehicles.agg({"Left": "max"}).collect()
```

```
# get min Left value
```

```
vehicles.agg({"Left": "min"}).collect()
```
